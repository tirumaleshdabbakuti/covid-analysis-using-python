#Analyzing the data

```python
total_cases = covid_df.new_cases.sum()
total_cases

total_deaths = covid_df.new_deaths.sum()
total_deaths, total_cases
```

```python
death_rate = covid_df.new_deaths.sum()/covid_df.new_cases.sum()
print(round(death_rate,4)*100,"%")
```

NAN values will be ignored during aggregating functions like sum()

```python
initial_tests = 935310
total_tests = initial_tests + covid_df.new_tests.sum()
total_tests
```

```python
positive_rate = total_cases/total_tests
print(f"{round(positive_rate,4)*100}% of tests in Italy led to a positive diagnosis.")
```

```python
high_cases = covid_df.new_cases > 1000
high_cases
```
